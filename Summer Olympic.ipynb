# Exploratory Data Analysis

```python
identify_team_events = pd.pivot_table(df_complete,
                                     index = ['Team','Edition','Event'],
                                     columns = 'Medal',
                                     values = 'Medal_Won',
                                     aggfunc = 'sum').reset_index()
identify_team_events = identify_team_events.loc[identify_team_events['Gold'] > 1, :]
identify_team_events
```

```python
team_sports = identify_team_events['Event'].unique()
team_sports
```

```python
import numpy as np
team_event_mask = df_complete['Event'].map(lambda x : x in team_sports)
single_event_mask = [not i for i in team_event_mask]

medal_mask = df_complete['Medal_Won'] == 1

df_complete['Team_Event'] = np.where(team_event_mask & medal_mask, 1, 0)
df_complete['Single_Event'] = np.where(single_event_mask & medal_mask, 1, 0)
df_complete['Event_Category'] = df_complete['Single_Event'] + df_complete['Team_Event']
```

```python
df_complete
```

```python
medal_tally_agnostic = df_complete.groupby(['Edition','Team','Event','Medal'])[['Medal_Won','Event_Category']].agg('sum').reset_index()

medal_tally_agnostic['Medal_Won_Corrected'] = medal_tally_agnostic['Medal_Won']/medal_tally_agnostic['Event_Category']
```

```python
medal_tally = medal_tally_agnostic.groupby(['Edition','Team'])['Medal_Won_Corrected'].agg('sum').reset_index()

medal_tally_pivot = pd.pivot_table(medal_tally,
                                  index = 'Team',
                                  columns = 'Edition',
                                  values = 'Medal_Won_Corrected',
                                  aggfunc = 'sum',
                                  margins=True).sort_values('All',ascending=False)[0:5]
medal_tally_pivot.loc[:,'All']
```

```python
import matplotlib.pyplot as plt
top_countries = ['United States','Soviet Union','United Kingdom','France']

year_team_medals = pd.pivot_table(medal_tally,
                                 index = 'Edition',
                                 columns = 'Team',
                                 values = 'Medal_Won_Corrected',
                                 aggfunc = 'sum')[top_countries]
year_team_medals.plot(linestyle = '-', marker = 'o', alpha = 0.9, figsize = (10,8), linewidth = 2)
plt.xlabel('Olympic Year')
plt.ylabel('Number of Medals')
plt.title('Olympic Perfomance Comparison')
```

```python
top_countries = ['United States','Soviet Union','United Kingdom','France']
row_mask2 = medal_tally_agnostic['Team'].map(lambda x: x in top_countries)

medal_tally_specific = pd.pivot_table(medal_tally_agnostic[row_mask2],
                                     index = ['Team'],
                                     columns = 'Medal',
                                     values = 'Medal_Won_Corrected',
                                     aggfunc = 'sum')
medal_tally_specific = medal_tally_specific.loc[:,['Gold','Silver','Bronze']]
medal_tally_specific.plot(kind = 'bar', stacked=True, figsize = (8,6), rot = 0)
plt.xlabel('Number of Medals')
plt.ylabel('Country')
```

```python
best_team_sports = pd.pivot_table(medal_tally_agnostic[row_mask2],
                                 index = ['Team','Event'],
                                 columns = 'Medal',
                                 values = 'Medal_Won_Corrected',
                                 aggfunc = 'sum').sort_values(['Team','Gold'], ascending = [True,False]).reset_index()
best_team_sports.drop(['Bronze','Silver'], axis = 1, inplace = True)
best_team_sports.columns = ['Team','Event','Gold_Medal_Count']
best_team_sports.groupby('Team').head(5)
```

# Do teams enjoy a home advantage?

```python
year_host_team = df_complete[['Edition','Country','Team']].drop_duplicates()
year_host_team
```

```python
row_mask4 = (year_host_team['Country'] == year_host_team['Team'])
year_host_team['Prev_Edition'] = year_host_team['Edition'] - 4
year_host_team['Next_Edition'] = year_host_team['Edition'] + 4

year_host_team = year_host_team[row_mask4]
year_host_team_medal = year_host_team.merge(medal_tally,
                                           left_on = ['Edition','Team'],
                                           right_on = ['Edition','Team'],
                                           how = 'left')
year_host_team_medal.rename(columns = {'Medal_Won_Corrected':'Medal_Won_Host_Year'}, inplace = True)

year_host_team_medal = year_host_team_medal.merge(medal_tally,
                                                 left_on = ['Prev_Edition','Team'],
                                                 right_on = ['Edition','Team'],
                                                 how = 'left')
year_host_team_medal.drop('Edition_y',axis=1,inplace=True)
year_host_team_medal.rename(columns = {'Medal_Won_Corrected':'Medal_Won_Prev_Year', 'Edition_x':'Edition'}, inplace = True)

year_host_team_medal = year_host_team_medal.merge(medal_tally,
                                                 left_on = ['Next_Edition','Team'],
                                                 right_on = ['Edition','Team'],
                                                 how = 'left')
year_host_team_medal.drop('Edition_y', axis=1,inplace=True)
year_host_team_medal.rename(columns = {'Medal_Won_Corrected':'Medal_Won_Next_Year', 'Edition_x':'Edition'}, inplace = True)

year_host_team_medal.drop(['Prev_Edition','Next_Edition'], axis=1,inplace=True)
year_host_team_medal.sort_values('Edition',ascending = True, inplace = True)
year_host_team_medal.reset_index(drop = True, inplace = True)

year_host_team_medal = year_host_team_medal.loc[:, ['Edition','Country','Team','Medal_Won_Prev_Year','Medal_Won_Host_Year','Medal_Won_Next_Year']]
year_host_team_medal
```
